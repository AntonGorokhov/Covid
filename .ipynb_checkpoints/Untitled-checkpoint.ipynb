```python
import pandas as pd
```

```python
df = pd.read_csv('owid-covid-data.csv', delimiter=',')
```

```python
df
```

```python
df = df[['date', 'location', 'total_cases_per_million', 'new_cases_per_million', 'new_cases_per_million']]
```

```python
import seaborn as sns
import matplotlib as plt
```

```python
df.shape
```

```python
locations = list(df['location'].unique())
len(locations)
```

```python
locations_2 = []
for i in locations:
    if (df[df.location == i]['total_cases_per_million'].count() >= 750):
        locations_2.append(i)
locations = locations_2
```

```python
locations
```

```python
df2 = df[df.location.isin(locations)]
```

```python
from tslearn.clustering import TimeSeriesKMeans
```

```python
from datetime import datetime
from sklearn.cluster import KMeans
import numpy as np
df2.head()
```

```python
df2['date'] = df2['date'].apply(lambda x : datetime.fromisoformat(x).timestamp())
```

```python
df2 = df2[['date', 'location', 'total_cases_per_million']]
df2.head()
```

```python
df2 = df2.dropna()
```

```python
df3 = np.array([df2[df2.location == x].drop(['location'], axis=1) for x in locations])
```

```python
df3[0]
```

```python
model = TimeSeriesKMeans(n_clusters=2, metric="dtw", max_iter=1, n_jobs=-1)
```

```python
dates_intersect = df3[0]['date'].unique()
for i in range(len(df3)):
    dates_intersect = list(set(dates_intersect) & set(df3[i]['date'].unique()))
for i in range(len(df3)):
    df3[i] = df3[i][df3[i].date.isin(dates_intersect)]
```

```python
len(dates_intersect)
```

```python
for i in range(len(df3)):
    df3[i] = df3[i].drop(['date'], axis=1).to_numpy() 
```

```python
df3[0].shape
```

```python
df3 = np.stack(df3)
df3.shape
```

```python
pred = model.fit_predict(df3)
```
